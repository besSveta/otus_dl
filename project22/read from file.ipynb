```python
import re
import os
replace_map = {}
replace_map['\n'] = ' '
replace_map['ӕ'] = 'æ'
replace_map['’'] = '\''
replace_map['’'] = '\''
replace_map['.'] = ' .'
replace_map[','] = ' ,'
replace_map[','] = ' ,'
replace_map['!'] = ' !'
replace_map['?'] = ' ?'
replace_map[':'] = ' :'
replace_map[';'] = ' ;'
replace_map['"'] = ' " '
replace_map['  '] = ' '
replace_map['\ufeff'] = ''

replace_map['ииа'] = 'ина'
replace_map['иио'] = 'ино'
replace_map['ииц'] = 'ниц'
replace_map['иид'] = 'нид'
replace_map['рии'] = 'рин'
replace_map['иин'] = 'инн'
replace_map['ииӕ'] = 'инæ'
replace_map['æии'] = 'æни'
replace_map['зииæ'] = 'зинæ'
replace_map['ьщц'] = 'ынц'
replace_map['ьщм'] = 'ынм'
replace_map['уьщи'] = 'уыцы'
replace_map['æиид'] = 'æнид'
replace_map['иич'] = 'нич'
replace_map['оии'] = 'они'
replace_map['иии'] = 'ини'
replace_map['иик'] = 'ник'
replace_map['иие'] = 'ине'


punctuation = '!"#$%&()*+-.,/:;<=>?@[\\]^_`{|}~»««»–№'
def ProcessText(text):
    if not re.match('.*[æа-яА-Я].*',text):
        return ''
    text = re.sub('[»««»„„]','"', text)
    for k,v in replace_map.items():
        text = text.replace(k,v)    
    text = re.sub(r'[a-zA-Z]|[#$%&§@<=>*#№\/]|[\\^_`{|}~»««»\[\]]|\[\d+\]', '', text)
    text = text.lower()
    if  re.match('.*[^æа-яА-Я\s0-9!\?\.,;\:()\-—"\'].*',text):
        return ''
    start_indx = 0
    if len(text) > 0:
        while start_indx < len(text) and (not re.match('[а-яА-Я]|[0-9]|[æ\-–"]',text[start_indx])):
            start_indx = start_indx + 1
        text = text[start_indx:]
    return text.strip()
```

```python
def get_sentences(text, splitter='.'):
    if (('.' not in text) and ('!' not in text) and ('?' not in text)):
        return [text]
    splitted = text.split(splitter)
    sentence = ''
    sentences = []

    for sent_part in splitted:
        if len(sent_part) < 3:
            sentence += sent_part + splitter
            continue
        if sentence !='':
            sent_part = sentence + sent_part
            sentence = ''
       
        if '?' in sent_part:
            qsplit = sent_part.split('?')
            for qs in qsplit:
                if qs!=qsplit[-1]:
                    qs = qs + '?'
                if '!' in qs:
                    hsplit = qs.split('!')
                    for hs in hsplit:
                        if hs!=hsplit[-1]:
                            hs = hs + '!'
                        if len(hs) > 2:
                            sentences.append(hs)
                elif len(qs)>2:
                        sentences.append(qs if qs != qsplit[-1] else qs + '.')
        elif '!' in sent_part:
            hsplit = sent_part.split('!')
            for hs in hsplit:
                if hs!=hsplit[-1]:
                    hs = hs + '!'
                if len(hs) > 2:
                    sentences.append(hs if hs != hsplit[-1] else hs + '.')
        elif len(sent_part) > 2:
            sentences.append(sent_part + '.')
        
    
    if sentence != '' and sentence not in sentences:
        sentences.append(sentence)
    

    return sentences
```
